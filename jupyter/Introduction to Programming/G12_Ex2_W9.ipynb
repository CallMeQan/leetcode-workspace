```python
# prompt: Create a class of Student with Name and GPA. Sort a list of students that the number of students and the data of each student are all inputted their data from the keyboard. A menu to select sorting the list by name or GPA are required. From the menu, select C to repeat the selection and E to exit.

from typing import List, Literal


class Student:
    def __init__(self, name: str, gpa: float):
        self.name = name
        self.gpa = gpa

    def __repr__(self):
        return f"Name: {self.name}, GPA: {self.gpa}"


def sort_students(students: List[Student], sort_by: Literal['name', 'gpa'] = 'name'):
    if sort_by == 'name':
        return sorted(students, key=lambda student: student.name)
    elif sort_by == 'gpa':
        return sorted(students, key=lambda student: student.gpa, reverse=True)
    else:
        return students

# Handle the user input, because user maybe want to exit the program
def exit_signal(input: any):
    if type(input) == str and input.lower() in ["e", "exit"]:
        return True
    elif type(input) == int and input == -1:
        return True
    return False

```

```python
num_students = int(input("Enter the number of students: "))
flag = True
if exit_signal(num_students):
    flag = False
students = []
for _ in range(num_students):
    name = input("Enter student name: ")
    if exit_signal(name):
        flag = False
        break
    gpa = float(input("Enter student GPA: "))
    if exit_signal(gpa):
        flag = False
        break
    students.append(Student(name, gpa))


while flag:
    print("\nMenu:")
    print("1. Sort by name")
    print("2. Sort by GPA")
    print("E. Exit")

    choice = input("Enter your choice: ").upper()

    if choice == '1':
        sorted_students = sort_students(students, 'name')
        print("\nSorted students by name:")
        for student in sorted_students:
            print(student)
    elif choice == '2':
        sorted_students = sort_students(students, 'gpa')
        print("\nSorted students by GPA:")
        for student in sorted_students:
            print(student)
    elif choice == 'E':
        flag = False
    else:
        print("Invalid choice. Please try again.")
```
