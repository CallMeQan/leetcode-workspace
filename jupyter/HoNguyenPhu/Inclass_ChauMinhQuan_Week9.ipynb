```python
### Exercise 1: counting the mid point
import random

class Point:
  def __init__(self, x = None, y = None):
      """
      Input x and y; or create a random point.

      If either x or y, or both are not given, a random value will be given.
      """
      # If both x and y are not given
      if x == None and y == None:
        self.x = random.randint(1, 10)
        self.y = random.randint(1, 10)

      # If only y is given
      elif x == None:
        self.x = random.randint(1, 10)
        self.y = y

      # If only x is given
      elif y == None:
        self.x = x
        self.y = random.randint(1, 10)

      # If both x and y are given
      else:
        self.x = x
        self.y = y

  def __str__(self):
      return "(" + str(int(self.x)) + ", " + str(int(self.y)) + ")"

  @staticmethod
  def midPoint(self, other):
      """
      Return the mid point of two points.
      """
      xm = (self.x + other.x) / 2
      ym = (self.y + other.y) / 2
      return Point(xm, ym)

  @staticmethod
  def Dist(self, other):
      """
      Return the distance of two points.
      """
      dx = self.x - other.x
      dy = self.y - other.y
      return (dx**2 + dy**2)**(1/2)

if __name__ == "__main__":
    # Initialize a and b
    a = Point(0, 0)
    b = Point(1, 0)

    # Print the points
    print(f"a = {a}")
    print(f"b = {b}")

    # Counting the mid point and the distance between a and b; print 'em
    mid = Point.midPoint(a, b)
    distance = Point.Dist(a, b)
    print(f"The middle point between a and b is: {mid}")
    print(f"The distance between a and b is: {distance}")
```

```python
### Exercise 2: counting the mid point
import random
import matplotlib.pyplot as plt

class Point:
    def __init__(self, x = None, y = None, max_range = 300):
        """
        Input x and y; or create a random point.

        If either x or y, or both are not given, a random value will be given.
        """
        # If both x and y are not given
        if x == None and y == None:
          self.x = random.randint(0, max_range)
          self.y = random.randint(0, max_range)

        # If only y is given
        elif x == None:
          self.x = random.randint(0, max_range)
          self.y = y

        # If only x is given
        elif y == None:
          self.x = x
          self.y = random.randint(0, max_range)

        # If both x and y are given
        else:
          self.x = x
          self.y = y

    def __str__(self):
        return "(" + str(int(self.x)) + ", " + str(int(self.y)) + ")"

    def __getitem__(self, indx):
        point = [self.x, self.y]
        return point[indx]

    def midPoint(self, other):
        """
        Arguments:

        :param self: A point
        :type self: instance
        :param other: Another point
        :type other: instance
        :return: The middle point of 2 points
        :rtype: instance of class Point
        """
        xm = (self.x + other.x) / 2
        ym = (self.y + other.y) / 2
        return Point(xm, ym)

    def distance(self, other):
        """
        Arguments:

        :param self: A point
        :type self: instance
        :param other: Another point
        :type other: instance
        :return: Distance between 2 points
        :rtype: float
        """
        dx = self.x - other.x
        dy = self.y - other.y
        return (dx**2 + dy**2)**(1/2)

class Disk:
    def __init__(self, x = None, y = None, r = 20, max_range = 300):
        """
        A class of disk, which has a center and a radius

        Methods include: intersects

        Arguments:

        :param r: The radius of the disk
        :type r: int
        :param max_range: The maximum range of the point
        :type max_range: int
        """
        self.center = Point(x, y, max_range = max_range)
        self.r = r

    def __str__(self):
        return str(self.center)

    @staticmethod
    def intersects(self, other):
        """
        Arguments:

        :param self: A list of disks
        :type self: list
        :param other: The radius of the circles
        :type other: int
        :return: Whether the circles intersect
        :rtype: bool
        """
        if Point.distance(self.center, other.center) < self.r + other.r:
            return True
        else:
            return False

def plot_circles(disks, radius, max_range):
    """
    Arguments:

    :param disks: A list of disks
    :type disks: list
    :param radius: The radius of the circles
    :type radius: int
    :param max_range: The maximum range of the point
    :type max_range: int
    :return: None
    :rtype: None
    """
    # Take coors
    x_coors = []
    y_coors = []
    circles = []
    for disk in disks:
        circles.append(plt.Circle((disk.center[0], disk.center[1]), radius))

    # Plot the circles (AI MADE - prompt: Plot the circles by matplotlib from my existing functions)
    fig, ax = plt.subplots()

    for circle in circles:
        ax.add_patch(circle)

    ax.set_aspect('equal')  # Ensure circles are circular
    ax.set_xlim([-21, max_range + 21])  # Set x-axis limits
    ax.set_ylim([-21, max_range + 21])  # Set y-axis limits
    plt.show()

if __name__ == "__main__":
    # Parameters
    disk_num = 20
    disks = []
    radius = 20
    max_range = 300
    num = 0

    # Loop until finish
    while num < disk_num:
        # Initialize new disk
        new_disk = Disk(x = None, y = None, r = radius, max_range = max_range)

        # The first disk will not be compared
        if num == 0:
            disks.append(new_disk)
            num += 1

        # Comparing the disks
        else:
            no_intersect = True
            for disk in disks:
                if Disk.intersects(new_disk, disk):
                    # print("Không thỏa mãn:")
                    # print("Hai điểm đang so sánh: ", new_disk.center, disk.center)
                    # print("Khoảng cách giữa hai điểm: ", Point.Dist(new_disk.center, disk.center), end = "\n\n")
                    no_intersect = False
                    break
            if no_intersect:
                disks.append(new_disk)
                num += 1

    # Print points
    for i in range(len(disks)):
        if i != 0:
            print(", ", end = "")
        print(disks[i], end = "")

    # Plot circles
    plot_circles(disks, radius, max_range)
```

```python
### Homework
from google.colab import output

# Tạo class di chuyển bằng hình tròn
class moving_Circle(Disk):
    def __init__(self, x = 0, y = 150, r = 10, gap = 5, max_range = 300):
        """
        Tạo ra đường tròn có thể di chuyển.

        Các giá trị khởi tạo đã ổn.
        """
        # Gán giá trị cho self
        self.x = x
        self.y = y
        self.r = r
        self.gap = gap
        self.max_range = max_range

        # Tạo đường tròn đầu tiên
        self.circles = []
        self.circles.append(Point(x, y, max_range))
        self.distance = r + gap

        # Khởi tạo các giá trị tính toán
        self.total = 1
        self.fail_num = 0
        self.best = self.x

        # Tạo giá trị x, y "hiện tại"
        self.current_x = self.x
        self.current_y = self.y

    def __getitem__(self, index):
        return self.circles[index]

    def __len__(self):
        return len(self.circles)

    def __str__(self):
        return str(self.circles)

    def move(self):
        """
        Di chuyển đường tròn. Sao cho:
        - Cách nhau khoảng cách 5
        - Nằm trong phạm vi
        """
        found = False
        while found == False:
            # Random giá trị x
            if self.current_x + self.distance < self.max_range:
                new_x = random.randint(self.current_x, self.current_x + self.distance)
            else:
                new_x = self.max_range

            # Dựa vào x để tính giá trị y
            new_y = abs(((self.distance)**2 - (self.current_x - new_x)**2)**(1/2) - self.current_y) # đã simplified

            # Kiểm tra điều kiện
            if self.current_x <= new_x <= self.max_range and 0 <= new_y <= self.max_range:
                found = True
                self.total += 1
                self.current_x = new_x
                self.current_y = new_y
                self.circles.append(Point(new_x, new_y, self.max_range))
            else:
                self.total += 1
                self.fail_num += 1
                if self.fail_num > 100:
                    found = True
                    print("FAil!")

    def is_not_there(self):
        """
        Kiểm tra xem đã đến đích chưa.
        """
        if self.current_x != self.max_range:
            return True
        else:
            return False

def plot_circles(disks, radius = 10, max_range = 300, delay = 0.1):
    """
    Arguments:

    :param disks: A list of disks
    :type disks: list
    :param radius: The radius of the circles
    :type radius: int
    :param max_range: The maximum range of the point
    :type max_range: int
    :return: None
    :rtype: None
    """
    # Take coors
    x_coors = []
    y_coors = []
    circles = []
    for disk in disks:
        circles.append(plt.Circle((disk[0], disk[1]), radius))

    # Plot the circles (AI MADE - prompt: Plot the circles by matplotlib from my existing functions)
    for circle in circles:
        output.clear()
        fig, ax = plt.subplots()
        ax.add_patch(circle)
        ax.set_aspect('equal')  # Ensure circles are circular
        ax.set_xlim([-21, max_range + 21])  # Set x-axis limits
        ax.set_ylim([-21, max_range + 21])  # Set y-axis limits
        plt.show()
        plt.pause(delay)

# if __name__ == "__main__":
def main(x = 0, y = 150, r = 10, gap = 5, max_range = 300, apply_plot = False, apply_print = False):
    # Parameters
    trontron = moving_Circle(x = x, y = y, r = r, gap = gap, max_range = max_range)

    while trontron.is_not_there():
        trontron.move()

    if apply_print == True:
        # Print kết quả
        for i in range(len(trontron)):
            if i != 0:
                print(", ", end = "")
            print(trontron[i], end = "")
    if apply_plot == True:
        # Plot điểm
        plot_circles(trontron)

    return trontron.total, trontron.fail_num, trontron.best
```
