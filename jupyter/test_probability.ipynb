```python
from random import random as rand
from math import sqrt
from typing import List, Tuple
```

```python
def random_mutiple():
    return [rand(), rand(), rand()]
```

```python
bruh: List[Tuple[float]] = [()]
for i in range(30):
    res = random_mutiple()
    x1 = res[0]
    x2 = res[1]
    x = res[2]
    bruh.append((max(x1, x2), sqrt(x)))

for i in range(30):
    print(bruh[i])
```
