```python
import csv

years: list = []
country_name = []
country_code = []
indicator_name = []
indicator_code = []

with open('data.csv', 'r') as file:
    reader = list(csv.reader(file))
    # Skip first 5 row, start read from row 6
    number_row = len(reader)
    number_col = len(reader[5])

    for row_index in range(5, number_row):
        tmp_years = []
        for col_index in range(number_col):
            if col_index == 0:
                country_name.append(reader[row_index][col_index])
            elif col_index == 1:
                country_code.append(reader[row_index][col_index])
            elif col_index == 2:
                indicator_name.append(reader[row_index][col_index])
            elif col_index == 3:
                indicator_code.append(reader[row_index][col_index])
            else:
                tmp_years.append(reader[row_index][col_index])
        years.append(tmp_years)

for i in range(69):
    print(country_name[i], country_code[i], indicator_name[i], indicator_code[i], years[i])
```
