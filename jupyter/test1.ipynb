```python
"""
The code appends the elements of the list t to the list x. but this time doing without append or extend method.
"""

def append_elements(x, t):
    for i in t:
        x += [i]
    return x
x = [1, 2, 3]
t = [4, 5, 6]
print(append_elements(x, t)) # [1, 2, 3, 4, 5, 6]
```

```python
"""
    Delete a element from a list without using the remove method.
"""
def delete_element(x, i):
    return [x[j] for j in range(len(x)) if j != i]
x = [1, 2, 3, 4, 5]
i_want_to_delete_the_number = 2
print(delete_element(x, i_want_to_delete_the_number)) # [1, 2, 4, 5]
```

```python
"""
    user input number n and position i, add n to the list x at position i without using the insert method.
"""
def add_element(x, n, i):
    return x[:i] + [n] + x[i:]
x = [1, 2, 3, 4, 5]
n = 10
i = 2
print(add_element(x, n, i)) # [1, 2, 10, 3, 4, 5]
```

```python
import random

def heapify(arr, n, i):
    largest = i
    l = 2 * i + 1
    r = 2 * i + 2

    if l < n and arr[i] < arr[l]:
        largest = l

    if r < n and arr[largest] < arr[r]:
        largest = r

    if largest != i:
        arr[i], arr[largest] = arr[largest], arr[i]
        heapify(arr, n, largest)

def heapsort(arr):
    n = len(arr)
    for i in range(n // 2 - 1, -1, -1):
        heapify(arr, n, i)
    for i in range(n-1, 0, -1):
        arr[i], arr[0] = arr[0], arr[i]
        heapify(arr, i, 0)
    return arr

def quicksort(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[len(arr) // 2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    return quicksort(left) + middle + quicksort(right)

# Generate a random array x with n values, each value is random from 1 to 100*n
n = int(input("Enter the number of elements: "))
x = [random.randint(1, 100 * n) for _ in range(n)]
print("Random array x:", x)

# Using Heapsort
sorted_x_heapsort = heapsort(x.copy())
print("Heapsort:", sorted_x_heapsort)

# Using QuickSort
sorted_x_quicksort = quicksort(x.copy())
print("QuickSort:", sorted_x_quicksort)
```

```python
import hashlib

# User input password
password = input("Enter your password: ")

# Encrypt the password using SHA-512
hashed_password = hashlib.sha512(password.encode()).hexdigest()
print("Encrypted password (SHA-512):", hashed_password)

# Decrypt the password using SHA-512
b = input("Enter the password to check: ")
if hashed_password == hashlib.sha512(b.encode()).hexdigest():
    print("Password is correct")
else:
    print("Password is incorrect")
```

```python
from cryptography.hazmat.primitives.asymmetric import rsa
from cryptography.hazmat.primitives import serialization

# Generate private key
private_key = rsa.generate_private_key(
    public_exponent=65537,
    key_size=1024,
)

# Generate public key
public_key = private_key.public_key()

# Serialize private key
private_pem = private_key.private_bytes(
    encoding=serialization.Encoding.PEM,
    format=serialization.PrivateFormat.TraditionalOpenSSL,
    encryption_algorithm=serialization.NoEncryption()
)

# Serialize public key
public_pem = public_key.public_bytes(
    encoding=serialization.Encoding.PEM,
    format=serialization.PublicFormat.SubjectPublicKeyInfo
)

print("Private Key:")
print(private_pem.decode('utf-8'))

print("Public Key:")
print(public_pem.decode('utf-8'))
```

```python
import numpy as np

# Generate two random 10x10 matrices
matrix_a = np.random.randint(1, 100, size=(5, 1))
matrix_b = np.random.randint(1, 100, size=(1, 5))

# Multiply the matrices
result_matrix = np.dot(matrix_a, matrix_b)

print("Matrix A:")
print(matrix_a)
print("\nMatrix B:")
print(matrix_b)
print("\nResult of multiplication:")
print(result_matrix)
```

```python
import matplotlib.pyplot as plt

# Flatten the result_matrix to a 1D array for easier plotting
flattened_result = result_matrix.flatten()

# Create a plot
plt.figure(figsize=(10, 6))
plt.plot(flattened_result, marker='o', linestyle='-', color='b')

# Add titles and labels
plt.title('Values in Result Matrix')
plt.xlabel('Index')
plt.ylabel('Value')

# Show the plot
plt.grid(True)
plt.show()
```

```python
import numpy as np
import time

# User input number n
n = int(input("Enter the number of elements: "))

# Linear Congruential Generator (LCG) parameters
a = 1664525
c = 1013904223
m = 2**32
seed = 42  # You can choose any seed value

# Generate array x with n length, each value is random from 1 to 100
x = []
for _ in range(n):
    seed = (a * seed + c) % m
    x.append(seed % 100 + 1)

print("Random array x:", x)
# Manually random based on current time from 1 to 100
current_time = int(time.time())
manual_random = (current_time % 100) + 1

print("Manual random number based on current time:", manual_random)
```
